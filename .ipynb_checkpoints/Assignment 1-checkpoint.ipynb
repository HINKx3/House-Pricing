# Data Preprocessing

```python
# Handle missing values and feature engineering for data_train and data_test

# Split the data_train dataset into features and target variable
X_train = train.drop('SalePrice', axis=1)
y_train = train['SalePrice']
```

# Model Selection and Splitting Data

```python

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_train, y_train, test_size=0.2, random_state=42)
```

```python


# Initialize and train the models
model_lr = LinearRegression()
model_lr = model_lr.fit(X_train, y_train)
model_rf = RandomForestRegressor(n_estimators=100)
# model_arima = ARIMA(y_train, order=(5,1,0)).fit()
```

# Model Evaluation

```python

# Linear Regression
y_pred_lr = model_lr.predict(X_test)
mae_lr = mean_absolute_error(y_test, y_pred_lr)
mse_lr = mean_squared_error(y_test, y_pred_lr)
rmse_lr = math.sqrt(mse_lr)

# Random Forest Regressor
y_pred_rf = model_rf.predict(X_test)
mae_rf = mean_absolute_error(y_test, y_pred_rf)
mse_rf = mean_squared_error(y_test, y_pred_rf)
rmse_rf = math.sqrt(mse_rf)

# ARIMA
#y_pred_arima = model_arima.forecast(steps=len(X_test))
#mse_arima = mean_squared_error(y_test, y_pred_arima)
#rmse_arima = math.sqrt(mse_arima)

print("Linear Regression - MAE: {}, RMSE: {}".format(mae_lr, rmse_lr))
print("Random Forest - MAE: {}, RMSE: {}".format(mae_rf, rmse_rf))
#print("ARIMA - MAE: {}, RMSE: {}".format(mae_arima, rmse_arima))
```
